```python
import pandas as pd
```

```python
alucar = pd.read_csv('alucar.csv')
alucar
```

```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
ax = sns.lineplot(x = 'mes', y = 'vendas', data = alucar)
ax.set_title('Vendas Mensais da Empresa', fontsize = 16)
ax.set_xlabel('Ms', fontsize = 14)
ax.set_ylabel('Vendas', fontsize = 14)
ax.figure.set_size_inches(12,6)
```

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar.head()
```

```python
ax = sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
ax.set_title('Aumento das Vendas Mensais da Empresa', fontsize = 16)
ax.set_xlabel('Ms', fontsize = 14)
ax.set_ylabel('Aumento das Vendas', fontsize = 14)
ax.figure.set_size_inches(12,6)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
ax = sns.lineplot(x = 'mes', y = 'aceleracao', data = alucar)
ax.set_title('Acelerao das Vendas Mensais da Empresa', fontsize = 16)
ax.set_xlabel('Ms', fontsize = 14)
ax.set_ylabel('Aumento acelerado das Vendas', fontsize = 14)
ax.figure.set_size_inches(12,6)
```
